```python
X = []
y = []
X_train = []
y_train = []
X_test = []
y_test = []
w = 56
l = 56
```

```python
fig = plt.figure(figsize=(10, 10))
for i in range(1, 6):
    img = cv2.imread(fr'inclusion/inclusion ({i}).jpg')
    img = cv2.resize(img, (56, 56))
    fig.add_subplot(2, 5, i)
    plt.imshow(img, aspect = 'auto')
    plt.grid(False)
for i in range(6,11):
    img = cv2.imread(fr'oil_spot/oil_spot ({i}).jpg')
    img = cv2.resize(img, (56, 56))
    fig.add_subplot(2, 5, i)
    plt.imshow(img, aspect = 'auto')
    plt.grid(False)
plt.show()

print(img.shape)
```

```python
for i in range(1, 516):
    img = cv2.imread(fr'inclusion/inclusion ({i}).jpg')
    img = cv2.resize(img, (l, w))
    X.append(img)
    y.append(0)

for i in range(1, 204):
    img = cv2.imread(fr'oil_spot/oil_spot ({i}).jpg')
    img = cv2.resize(img, (l, w))
    X.append(img)
    y.append(1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3)
```

```python
X_train = np.array(X_train, dtype = 'float')
y_train = np.array(y_train, dtype = 'float').astype('float32').reshape((-1,1))
X_test = np.array(X_test, dtype = 'float')
y_test = np.array(y_test, dtype = 'float').astype('float32').reshape((-1,1))
```

```python
n_classes = 2

X_train /= 255
X_test /= 255
```

```python
w=l=56
model = Sequential()
model.add(Conv2D(56, kernel_size=(2,2), strides=(1,1), padding='valid', activation='relu', input_shape=(w,l,3)))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(112, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(224, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(112, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(56, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(10, activation='linear'))

model.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001), metrics = ['accuracy'])
```

```python

model.summary()
```

```python
model.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(), metrics = ['accuracy'])
```

```python
H = model.fit(X_train, y_train, batch_size = 10, epochs = 30, validation_data = (X_test, y_test))
```

```python
N=30
plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, N), H.history["loss"], label="train_loss")
plt.plot(np.arange(0, N), H.history["accuracy"], label="train_acc")
plt.plot(np.arange(0, N), H.history["val_accuracy"], label="val_acc")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="lower left")
plt.show()
```
